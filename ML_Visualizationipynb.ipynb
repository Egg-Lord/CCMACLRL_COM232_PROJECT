```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


sns.set_style('whitegrid')
```

### Load Data

```python
# Replace the paths with your actual CSV paths in Colab
train = pd.read_csv('/content/Training.csv')
test = pd.read_csv('/content/Testing.csv')
train_cleaned = pd.read_csv('/content/Training_Cleaned.csv')
test_cleaned = pd.read_csv('/content/Testing_Cleaned.csv')
```

### Text Visualization


```python
# Show the first few rows to see the structure
print("Training dataset:")
display(train.head(10))

print("Testing dataset:")
display(test.head(10))

print("Training Cleaned dataset:")
display(train_cleaned.head(10))

print("Testing Cleaned dataset:")
display(test_cleaned.head(10))

# Show column names and types
print("Columns and types for Training:")
print(train.info())

print("Columns and types for Testing:")
print(test.info())

print("Columns and types for Training Cleaned:")
print(train_cleaned.info())

print("Columns and types for Testing Cleaned:")
print(test_cleaned.info())

print("Target distribution in Training:")
print(train['prognosis'].value_counts())

print("Target distribution in Training Cleaned:")
print(train_cleaned['prognosis'].value_counts())
```

### Target Distribution


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(y='prognosis', data=train, palette='viridis', order=train['prognosis'].value_counts().index)
plt.title("Disease Distribution in Training Set")
plt.xlabel("Count")
plt.ylabel("Disease")
plt.show()

plt.figure(figsize=(12,6))
sns.countplot(y='prognosis', data=train_cleaned, palette='magma', order=train_cleaned['prognosis'].value_counts().index)
plt.title("Disease Distribution in Training Cleaned Set")
plt.xlabel("Count")
plt.ylabel("Disease")
plt.show()
```

### Sympton Frequency


```python
symptoms = train.columns[:-2]  # exclude 'prognosis' and any unnamed columns

# Sum of symptoms for all patients
symptom_counts = train[symptoms].sum().sort_values(ascending=False)

plt.figure(figsize=(15,6))
sns.barplot(x=symptom_counts.index, y=symptom_counts.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title("Symptom Frequency in Training Set")
plt.ylabel("Number of Patients")
plt.show()
```

### Sympton Frequency


```python
# Example: top 5 symptoms for Fungal infection
disease = "Fungal infection"
top_symptoms = train[train['prognosis']==disease][symptoms].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,4))
sns.barplot(x=top_symptoms.index, y=top_symptoms.values, palette='Set2')
plt.title(f"Top 5 Symptoms for {disease}")
plt.ylabel("Count")
plt.show()
```

### Sympton Correlation


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix for numeric columns only
correlation =train_cleaned.corr(numeric_only=True)

# Create a figure
plt.figure(figsize=(16, 14))

# Draw a clean heatmap
sns.heatmap(
    correlation,
    annot=False,           # remove annotations for large matrices
    cmap="coolwarm",       # visually clear diverging colors
    vmin=-1, vmax=1, center=0,
    square=True,
    linewidths=0.3,
    cbar_kws={"shrink": 0.5, "label": "Correlation"}
)

# Tidy up the labels
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.title('Correlation Matrix', fontsize=18, pad=20)

# Save and show
plt.tight_layout()
plt.savefig('correlation_heatmap_clean.png', dpi=300)
plt.show()
```

```python
train['symptom_count'] = train[symptoms].sum(axis=1)
plt.figure(figsize=(12,6))
sns.boxplot(x='symptom_count', y='prognosis', data=train)
plt.title("Number of Symptoms per Disease")
plt.xlabel("Number of Symptoms")
plt.ylabel("Disease")
plt.show()
```
